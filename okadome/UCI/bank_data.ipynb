```python
import numpy as np
import sklearn
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
bank_data = pd.read_csv("bank/data_banknote_authentication.csv", header=None)
```

```python
len(bank_data)
```

```python
a = np.random.randint(0, len(bank_data), size=5)
bank_data.iloc[a].reset_index(drop=True)
```

```python
def closs_val_test(models, params, x, y):
    kf = KFold(n_splits=5, shuffle=True, random_state=2018)
    result = []
    predict_values = []
    for train, test in kf.split(x):
        grid = GridSearchCV(models, params, cv=5)
        search_result = grid.fit(x[train], y[train])
        best = search_result.best_estimator_
        output = best.predict(x[test])
        y_true = y[test]
        result.append(accuracy_score(y_true, output))
        predict_values.append(output*1)
        print ("best_model:", best)
        print ("output_values:", output)
        print ("true_values:", y_true)
        print ("accuracy:", accuracy_score(y_true, output))
    return result, predict_values
```

```python
models = SVC()
params = {'kernel':['rbf', 'linear'] ,'C': [10**i for i in range(-3, 3)],  'gamma' : [2**i for i in range(-5,5)]}
x = bank_data.iloc[:,:-1].values
y = bank_data.iloc[:,-1].values
ss_x = StandardScaler().fit(x)
x_trans = ss_x.transform(x)
```

```python
result_svc, values_svc = closs_val_test(models, params, x_trans, y)
```

```python
print ("SVR accuracy", np.mean(result_svc))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
models_l = LogisticRegression()
params = {'C': [10**i for i in range(-3, 3)]}
result, values = closs_val_test(models_l, params, x_trans, y)
```

```python
print ("Logistic accuracy", np.mean(result))
```

```python
from  sklearn.ensemble import RandomForestClassifier
model_f = RandomForestClassifier()
params = {'n_estimators': list(range(5,100,5)), 'min_samples_split' : list(range(2,10))}
result_f, values_f = closs_val_test(model_f, params, x_trans, y)
```

```python
print ("RandomForest accuracy", np.mean(result_f))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_k = KNeighborsClassifier()
params = {'n_neighbors': list(range(3,27,2))}
result_k, values_k = closs_val_test(model_k, params, x_trans, y)
```

```python
print ("kNN accuracy", np.mean(result_k))
```
