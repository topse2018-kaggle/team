### 5. データの尺度を正規化する

#### 正規化方法を検討するため現時点の説明変数の分布を表示

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.hist(df['X1'].values)
ax2.hist(df['X2'].values)
ax3.hist(df['X3'].values)
ax4.hist(df['X4'].values)

ax1.set_title('X1')
ax2.set_title('X2')
ax3.set_title('X3')
ax4.set_title('X4')

fig.tight_layout()
plt.show()
```

説明変数は全て標準正規分布で正規化する。

```python
X_df = df[['X1','X2','X3', 'X4']]
y_df = df['y']

df_np = StandardScaler().fit_transform(X_df)
X_df_std = pd.DataFrame(df_np, columns=['X1','X2','X3', 'X4'])

df_std = X_df_std.join(y_df)
df_std.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.hist(df_std['X1'].values)
ax2.hist(df_std['X2'].values)
ax3.hist(df_std['X3'].values)
ax4.hist(df_std['X4'].values)

ax1.set_title('X1')
ax2.set_title('X2')
ax3.set_title('X3')
ax4.set_title('X4')

fig.tight_layout()
plt.show()
```

## モデル生成・評価・予測

### モデル生成

#### 学習データと評価データに分割する

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_df_std,
    y_df, 
    test_size = 300,
    random_state = 0
    )
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

#### 交差検証とグリッドサーチを同時に実行する

```python
parameters = { 'C':np.logspace(-4,4,20), 'gamma':np.logspace(-4,4,20)}
grid_search = GridSearchCV(
                    SVC(kernel='rbf', probability=True), 
                    parameters, 
                    cv = 2
                )
grid_search.fit(X_train, y_train)
```

### モデル評価

#### スコアとモデルパラメータのベスト値

```python
print("訓練データの正解率(ベスト値): ", grid_search.best_score_)
print("ベスト時のSVCパラメータ:", grid_search.best_params_)
```

### モデル予測

```python
pred = grid_search.predict(X_test)
print("正解率 ＝ ", accuracy_score( y_test, pred))
print("適合率 ＝ ", precision_score( y_test, pred))
print("再現率 ＝ ", recall_score( y_test, pred))
print("F値 ＝ ", f1_score( y_test, pred))
print("混合行列 ＝ \n",confusion_matrix(y_test, pred))
```

#### 【おまけ】ROC曲線とAUC

```python
proba = grid_search.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, proba) 
```

```python
plt.plot(fpr, tpr)
plt.title("ROC curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

```python
auc(fpr, tpr)
```
